```python
import astropy.units as u

from astropy.coordinates import EarthLocation
from visibility import ObjectVisibility
```

```python
# observatory location definition
calar_alto = EarthLocation(lat = 37.2236*u.deg, lon = -2.5461*u.deg, height = 2168*u.m)
# spanish time offset the observation date (utc +1 in March)
utcoffset = 1*u.hour  
NGC5395 = ObjectVisibility('2020-3-10', utcoffset, calar_alto, 'NGC5395')
```

The best time to observe the galaxies with minimal sky background noise and lowest air-mass is during the time of maximum angular distance between them and the moon, which occurs between 1am and 6am. To ensure that the object is observable by the telescope, the time interval is delimited by vertical yellow lines, which indicate when the object is 30ยบ above the horizon.

```python
NGC5395.plot_alt(figsize=(9, 6))#, save=True)
```

The path followed by the object in the night sky the date of the observation. \
As seen in the image, the object rises from the North-West and sets North-East.
